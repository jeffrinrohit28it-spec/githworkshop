```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
blank_img = np.zeros((512,512,3), dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
plt.show
```

```python
cv2.rectangle(blank_img, pt1=(384,0), pt2=(510,128), color = (0,255,0), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.rectangle(blank_img, pt1=(200,200), pt2=(300,300), color = (0,0,255), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(blank_img, center=(100,100), radius=50,color=(255,0,0), thickness=5)
plt.imshow(blank_img)
plt.show()
```

```python
cv2.circle(blank_img, center=(250,250), radius=50,color=(255,0,0), thickness=5)
plt.imshow(blank_img)
plt.show()
```
